Generate a directed graph for a given site, where each edge stores underlying terrain properties.

## Notebook imports and logging setup

````python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import os
from pathlib import Path
import logging
import logging.config
import json
import yaml
import random
import ipynbname
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from gplanetary_nav.site.loader import load as site_loader
from gplanetary_nav.dev.paths_config import REPO_CFG as GPLANETARY_CFG

# Logging setup
with open(os.path.join(GPLANETARY_CFG.config_dir, 'logging.json')) as f:
    log_cfg = json.load(f)
lfn = os.path.join(GPLANETARY_CFG.logs_dir, ipynbname.name()+'.log')
log_cfg['handlers']['file']['filename'] = lfn
logging.config.dictConfig(log_cfg)
log = logging.getLogger(__name__)
print(f"Logging to {lfn}")

````

## Create graph with edge features

````python
# Load world config, add full path to dataset to load
world_fpath = Path(GPLANETARY_CFG.params_dir, "world.yaml")
with open(world_fpath) as f:
    world_cfg = yaml.load(f, Loader=yaml.FullLoader)
world_cfg['site_dirpath'] = Path(os.environ.get('GNAV_DATASET_PATH'), world_cfg['site_name'])

# Load rover config
rover_fpath = Path(GPLANETARY_CFG.params_dir,"rover.yaml")
with open(rover_fpath) as f:
    rover_cfg = yaml.load(f, Loader=yaml.FullLoader)
````

````python
# Create graph
from gplanetary_nav.graph.grid_graph import GridGraph

gg = GridGraph.load(world_cfg, rover_cfg)
G = gg.G     # G attribute is a NetworkX DiGraph

# Show number of nodes and edges in the graph
log.info(f"Num of nodes in graph: {len(G.nodes)}")
log.info(f"Num of edges in graph: {len(G.edges)}")

# Read one edge to see what its attributes are
for _, _, features_dict in G.edges(data=True):
    log.info(f"Graph edge attributes: {features_dict.keys()}")
    break

# Min/max of edge features across graph
for feature in features_dict.keys():
    min_val = min(dict(G.edges).items(), key=lambda x: x[1][feature])[1][feature]
    max_val = max(dict(G.edges).items(), key=lambda x: x[1][feature])[1][feature]
    log.info(f"Feature '{feature}': min {min_val:.2f}, max {max_val:.2f}")
````

## Find physically shortest path

````python
from gplanetary_nav.utils import path_cost

# Plan
# start_cell, goal_cell = random.choices(list(G.nodes), k=2)
start_cell, goal_cell = (176, 175), (18, 87)
path = nx.shortest_path(G, start_cell, goal_cell, weight='length')

path_length = path_cost(G, path, 'length')
path_duration = path_cost(G, path, 'time_total')

# Load site
dirpath = Path(os.environ.get('GNAV_DATASET_PATH'), "JEZ5_sample")
site = site_loader(dirpath)

fig, ax = plt.subplots(1,1)
ax = site.layers['dem'].plot(ax, cmap='bwr')
ax.imshow(site.nogo==0, extent=site.extent, cmap='gray', alpha=0.3)

utm_path = np.array([site.UTM_from_grid(n) for n in path])
X_plot = [site.X[cell] for cell in path]
Y_plot = [site.Y[cell] for cell in path] 
ax.plot(utm_path[:,0], utm_path[:,1], color='k', linewidth=1, label="Path")

ax.scatter(*site.UTM_from_grid(start_cell), c='y', edgecolor='k', label='Start', zorder=5)
ax.scatter(*site.UTM_from_grid(goal_cell), c='g', edgecolor='k', label='Goal',  zorder=5)

ax.legend()
ax.set_title(f"Physically shortest path ({path_length:.0f}m, {path_duration/3600:.1f}h) \n (Elevation & no-go map overlay)")

plt.savefig("/tmp/shortest_path.png", dpi=75, bbox_inches="tight")
````

## Planning with the MTTT velocity model

For this part, we load a dataset configuration using a different terrain layer/map and employ the M2020 Terrain Traversability analysis Tools (MTTT) velocity model.

Before continuing, modify the terrain layer parameters in the dataset's `settings.yaml` as follows:

```yaml
#[...]
    terrain:
        fpath: terrain_mttt.tif
        labels: !include terrain_labels_mttt.yaml
        avoid:
          - 4
          - 255
# [...]
```



````python
# Load world config with mttt velocity model
world_fpath = Path(GPLANETARY_CFG.params_dir, "world_mttt.yaml")
with open(world_fpath) as f:
    world_cfg = yaml.load(f, Loader=yaml.FullLoader)
world_cfg['site_dirpath'] = Path(os.environ.get('GNAV_DATASET_PATH'), world_cfg['site_name'])

# Load rover config
rover_fpath = Path(GPLANETARY_CFG.params_dir,"rover.yaml")
with open(rover_fpath) as f:
    rover_cfg = yaml.load(f, Loader=yaml.FullLoader)
    
# Generate grid graph
gg = GridGraph.load(world_cfg, rover_cfg)
G = gg.G
````

````python
path = nx.shortest_path(G, start_cell, goal_cell, weight='time_total')

path_length = path_cost(G, path, 'length')
path_duration = path_cost(G, path, 'time_total')

# Load site
dirpath = Path(os.environ.get('GNAV_DATASET_PATH'), "JEZ5_sample")
site = site_loader(dirpath)

fig, ax = plt.subplots(1,1)
ax = site.layers['terrain_mttt'].plot(ax, cmap='bwr')
ax.imshow(site.nogo==0, extent=site.extent, cmap='gray', alpha=0.5)

utm_path = np.array([site.UTM_from_grid(n) for n in path])
X_plot = [site.X[cell] for cell in path]
Y_plot = [site.Y[cell] for cell in path] 
ax.plot(utm_path[:,0], utm_path[:,1], color='k', linewidth=1, label="Path")

ax.scatter(*site.UTM_from_grid(start_cell), c='y', edgecolor='k', label='Start', zorder=5)
ax.scatter(*site.UTM_from_grid(goal_cell), c='g', edgecolor='k', label='Goal',  zorder=5)

ax.legend()
ax.set_title(f"Fastest path ({path_length:.0f}m, {path_duration/3600:.1f}h) \n (Terrain & no-go map overlay)")

plt.savefig("/tmp/shortest_path_mttt.png", dpi=75, bbox_inches="tight")
````
