# Random terrain generator demo
Scripts that generates random terrain elevation maps, slope and aspect maps, and exports them to a `.tif` format

Author: Olivier Lamarre  
Affl.: STARS Laboratory, University of Toronto

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import os
from pathlib import Path
import json
import time
import logging
import logging.config

import ipynbname
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

from gplanetary_nav.site.generator import SiteGenerator
from gplanetary_nav.dev.paths_config import REPO_CFG as GPLANETARY_CFG

# Logging setup
with open(os.path.join(GPLANETARY_CFG.config_dir, 'logging.json')) as f:
    log_cfg = json.load(f)
lfn = os.path.join(GPLANETARY_CFG.logs_dir, ipynbname.name()+'.log')
log_cfg['handlers']['file']['filename'] = lfn
logging.config.dictConfig(log_cfg)
log = logging.getLogger(__name__)
print(f"Logging to {lfn}")
```

## First, generate unitless elevation model with desired roughness

```python
# Print initialization parameters info
# help(SiteGenerator)

# Diamond Square method example
# roughness = 0.3
# sg = SiteGenerator((150,300), method="DS", roughness=roughness)

# Perlin noise method example
per = (5,10)
til=(False,False)
sg = SiteGenerator((150,300), method="PE", per=per, til=til)

# Plot terrain generated
fig = plt.figure(figsize=(12,5))
fig.tight_layout()

ax = fig.add_subplot(1, 1, 1)
im = ax.imshow(sg.raw_dem, extent=[0,1.0,sg.hwratio,0],cmap=cm.coolwarm)
fig.colorbar(im, ax=ax)

ax.set_xlabel("Easting (units)")
ax.set_ylabel("Northing (units)")
ax.set_title("Top view, (elevation in units)")
ax.set_xlim(0.0, 1.0)
ax.set_ylim(0.0, sg.hwratio)

plt.show()
```

## Second, adjust elevation scaling, and set resolution per pixel

```python
# Resolution (meters) per pixel
# (higher = shallower slopes)
res = 1.0

# Elevation scaling (meters per unit)
# (lower scale = shallower slopes)
scale = 3.0

# Create elevation, slope and aspect maps
(dem, slope, aspect) = sg.create_geometric_maps(
    res=res, scale=scale, tmp_dir=GPLANETARY_CFG.data_dir)

# Plot results
fig = plt.figure(figsize=(20,4))
fig.tight_layout()

# Show terrain, top view
ax = fig.add_subplot(1, 3, 1)
im = ax.imshow(dem, extent=sg.plotting_extent,cmap=cm.coolwarm)
fig.colorbar(im, ax=ax)

ax.set_xlabel("Width (meters)")
ax.set_ylabel("Height (meters)")
ax.set_title("Top view \n(elevation in meters)")
ax.set_xlim(0.0, sg.plotting_extent[1])
ax.set_ylim(0.0, sg.plotting_extent[2])

# Show slope map, top view
ax = fig.add_subplot(1, 3, 2)
im = ax.imshow(slope, extent=sg.plotting_extent,cmap=cm.coolwarm)
fig.colorbar(im, ax=ax)

ax.set_xlabel("Width (meters)")
ax.set_ylabel("Height (meters)")
ax.set_title("Top view \n(slope magnitude in degrees)")
ax.set_xlim(0.0, sg.plotting_extent[1])
ax.set_ylim(0.0, sg.plotting_extent[2])

# Show aspect map, top view
ax = fig.add_subplot(1, 3, 3)
im = ax.imshow(aspect, extent=sg.plotting_extent,cmap=cm.coolwarm)
fig.colorbar(im, ax=ax)

ax.set_xlabel("Width (meters)")
ax.set_ylabel("Height (meters)")
ax.set_title("Top view \n(aspect in degrees)")
ax.set_xlim(0.0, sg.plotting_extent[1])
ax.set_ylim(0.0, sg.plotting_extent[2])

plt.show()
```

## Create a terrain class map, if desired

```python
# Terrain class generation parameters
num_clusters = 100 # Number of elliptical terrain patches/clusters sampled
num_classes = 5  # Number of distinct terrain types
max_cluster_width = 20  # The max width of every cluster, in px

# Create terrain classes
terrain = sg.create_terrain_classes(num_clusters=num_clusters,
                                    num_classes=num_classes,
                                    max_cluster_width=max_cluster_width)

# Plot terrain types generated
fig = plt.figure(figsize=(12,5))
fig.tight_layout()

ax = fig.add_subplot(1, 1, 1)
im = ax.imshow(terrain, extent=sg.plotting_extent,cmap=cm.get_cmap('viridis', num_classes))
cbar = fig.colorbar(im, ax=ax, ticks=list(range(num_classes)))

# Center ticks on corresponding color in colorbar
tick_locs = (np.arange(num_classes)+0.5)*(num_classes-1)/num_classes
cbar.set_ticks(tick_locs)
cbar.set_ticklabels(np.arange(num_classes))

ax.set_xlabel("Width (meters)")
ax.set_ylabel("Height (meters)")
ax.set_title("Terrain classes")
ax.set_xlim(0.0, sg.plotting_extent[1])
ax.set_ylim(0.0, sg.plotting_extent[2])

plt.show()
```

## If maps look good, save them

```python
# Miscillaneous info about how the maps were generated (for reproducibility)

# Example for Perlin noise method
misc_info = f"Generated with PE, num. periods {per}, tiling {til}, resolution {res} m/px, scale {scale}, terrain clusters {num_clusters}, terrain cluster max width {max_cluster_width}"

# Example for Diamond Square method
# misc_info = "Generated with DS, roughness {}, resolution {} m/px, scale {}, terrain clusters {}, terrain cluster max width {} ".format(roughness, res, scale, num_clusters, max_cluster_width)

# Set desired config name (also the name of subdirectory)
save_dir = Path(os.environ.get('GNAV_DATASET_PATH'), f"dataset{int(time.time())}")
sg.save_dataset(save_dir=save_dir, misc_info=misc_info)
```
